```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

from lazypredict.Supervised import LazyRegressor

from sklearn.preprocessing import LabelEncoder, OneHotEncoder

from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("../data/processed/train.csv")
```

```python
train_set, test_set = train_test_split(df, test_size=0.2, random_state=42)
```

```python
X_train, X_val, y_train, y_val = train_test_split(train_set.drop(columns=["price"]), train_set[["price"]], test_size=0.2, random_state=42)
```

```python
X_train
```

```python
le = LabelEncoder()

X_train["color"] = le.fit_transform(X_train["color"])
X_train["province"] = le.fit_transform(X_train["province"])

X_val["color"] = le.fit_transform(X_val["color"])
X_val["province"] = le.fit_transform(X_val["province"])
```

```python
ohe = OneHotEncoder(sparse=False, handle_unknown='ignore')

ohe.fit(X_train[['make']])

transformed_X_train = ohe.transform(X_train[['make']])
transformed_df = pd.DataFrame(transformed_X_train, columns=ohe.get_feature_names_out(['make']), index=X_train.index)
X_train_ohe = pd.concat([X_train, transformed_df], axis=1).drop(columns='make')

transformed_X_val = ohe.transform(X_val[['make']])
transformed_df = pd.DataFrame(transformed_X_val, columns=ohe.get_feature_names_out(['make']), index=X_val.index)
X_test_ohe = pd.concat([X_val, transformed_df], axis=1).drop(columns='make')
```

```python
X_train_ohe["year"] = X_train_ohe["year"].astype("int16")
X_train_ohe["power"] = X_train_ohe["power"].astype("int16")
X_train_ohe["is_professional"] = X_train_ohe["is_professional"].astype("int8")

X_test_ohe["year"] = X_test_ohe["year"].astype("int16")
X_test_ohe["power"] = X_test_ohe["power"].astype("int16")
X_test_ohe["is_professional"] = X_test_ohe["is_professional"].astype("int8")
```

```python
# Selecciona las columnas de tipo float
columnas_float = [columna for columna in X_train_ohe.columns if X_train_ohe[columna].dtype == "float64"]

# Convierte las columnas a tipo int
for columna in columnas_float:
    X_train_ohe[columna] = X_train_ohe[columna].astype("int8")
    X_test_ohe[columna] = X_test_ohe[columna].astype("int8")
```
